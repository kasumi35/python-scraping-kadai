```python
# 検索キーワードを変数search_wordに格納
station_name = input('検索キーワードを入力してください: ')
```

```python
# Geocoding APIのURL
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'key': api_key,
        'address': station_name + '駅'
    }
)


```

```python
# レスポンスのJSONを取得
json_data = response.json()

# JSONの中身を確認
pprint.pprint(json_data)
```

```python
# 取得した動画のタイトルとURLを表示
for item in json_data['results']:
    #緯度
    latitude_data = item['geometry']['location']['lat']
    print(f"緯度:  {latitude_data}")

#    print("緯度"+ latitude )
```

```python
# 取得した動画のタイトルとURLを表示
for item in json_data['results']:
    #経度
    longitude_data = item['geometry']['location']['lng']
    print(f"経度:  {longitude_data}")
```
